Ejercicio 1: Análisis de Transacciones de E-commerce

Carga el archivo examen_transacciones.xlsx (2 punto)
2. Visualiza los primeros 10 datos de la tabla (3 puntos)
3. Imprime el nombre de las variables (3 puntos)
4. Crea una columna llamada "Subtotal" que multiplique Cantidad por Precio_Unitario (3 puntos).
5. Crea una columna a la derecha de “Subtotal” llamada “IVA Trasladado” que multiplique el subtotal por 0.16 (3 puntos)


```python
import pandas as pd

# Carga el archivo examen_transacciones.xlsx (2 punto)
df = pd.read_excel('examen_transacciones.xlsx')

# Visualiza los primeros 10 datos de la tabla (3 puntos)
print(df.head(10))
print('')

# Imprime el nombre de las variables (3 puntos)
print(df.columns)
print('')

# Crea una columna llamada "Subtotal" que multiplique Cantidad por Precio_Unitario (3 puntos).
df['Subtotal'] = df['Cantidad'] * df['Precio_Unitario']
print('Subtotal: ', df['Subtotal'])
print('')

# Crea una columna a la derecha de “Subtotal” llamada “IVA Trasladado” que multiplique el subtotal por 0.16 (3 puntos)
df['IVA Trasladado'] = df['Subtotal'] * 0.16
print('IVA Trasladado: ', df['IVA Trasladado'])

```

Filtra solo las transacciones con estado "Completado" (2 puntos)
7. Calcula el total de ventas completadas por producto incluyendo impuestos (3 puntos)
8. Identifica cuál fue el total de las ventas con IVA incluido e imprime el resultado en salida estándar creando una variable llamada “Total_Ventas” (3 puntos)
9. Calcula el máximo de ventas por cliente (considera solo transacciones completadas) (3 puntos)
10. Crea una columna llamada “Max_ventas_clitente” que guarde el máximo de ventas para cada cliente calculado en el paso 9 (3 puntos)


```python
# Filtra solo las transacciones con estado "Completado" (2 puntos)
df_completados = df[df['Estado'] == 'Completado'].copy()
print(df_completados)

#  Calcula el total de ventas completadas por producto incluyendo impuestos (3 puntos)
df_completados['Total_Ventas'] = df_completados['Subtotal'] + df_completados['IVA Trasladado']
print(df_completados)

# 8. Identifica cuál fue el total de las ventas con IVA incluido e imprime el resultado en salida estándar creando una variable llamada “Total_Ventas” (3 puntos)
total_ventas = df_completados['Total_Ventas'].sum()
print('Total de ventas con IVA incluido: ', total_ventas)

# Calcula el máximo de ventas por cliente (considera solo transacciones completadas) (3 puntos)
max_ventas_por_cliente = df_completados.groupby('Cliente_ID')['Total_Ventas'].max()
print(max_ventas_por_cliente)

# 10. Crea una columna llamada “Max_ventas_clitente” que guarde el máximo de ventas para cada cliente calculado en el paso 9 (3 puntos)
df_completados['Max_ventas_cliente'] = df_completados['Cliente_ID'].map(max_ventas_por_cliente)
print(df_completados)
```

Calcula el promedio de ventas por producto (considera solo transacciones completadas) (3 puntos)
12. Crea una columna llamada “Mean_ventas_producto” que guarde el promedio de ventas para cada cliente calculado en el paso 11 (3 puntos)
13. Guarda el resumen de ventas por producto en un archivo llamado "Ventas_completadas_[tus iniciales].xlsx" por ejemplo, “Ventas_completadas_SLFB” (3 puntos)
14. Adjunta la base de datos a la entrega (3 puntos)


```python
# Calcula el promedio de ventas por producto (considera solo transacciones completadas) (3 puntos)
mean_ventas_por_producto = df_completados.groupby('Producto')['Total_Ventas'].mean()
print(mean_ventas_por_producto)

# Crea una columna llamada “Mean_ventas_producto” que guarde el promedio de ventas para cada cliente calculado en el paso 11 (3 puntos)
df_completados['Mean_ventas_producto'] = df_completados['Producto'].map(mean_ventas_por_producto)
print(df_completados)

# Guarda el resumen de ventas por producto en un archivo llamado "Ventas_completadas_[tus iniciales].xlsx" por ejemplo, “Ventas_completadas_SLFB” (3 puntos)
df_resumen_ventas = df_completados.groupby('Producto')['Total_Ventas'].sum().reset_index()
df_resumen_ventas.to_excel('Ventas_completadas_JAMP.xlsx', index=False)
```
